```python
import nltk
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
texts = "Hi this is manoj"
for text in texts:
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        print(words)
        tagged = nltk.pos_tag(words)
        print(tagged)
```
